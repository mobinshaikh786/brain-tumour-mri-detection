```python
import os, matplotlib.pyplot as plt
from PIL import Image
DATA = 'dataset_sample'
for cls in os.listdir(DATA):
    p = os.path.join(DATA, cls)
    if os.path.isdir(p):
        print(f'{cls}: {len(os.listdir(p))} images')
```

```python
# Show 1 sample per class
fig, axes = plt.subplots(1, 4, figsize=(14,4))
for ax, cls in zip(axes, sorted(os.listdir(DATA))):
    p = os.path.join(DATA, cls)
    if os.path.isdir(p) and os.listdir(p):
        img = Image.open(os.path.join(p, os.listdir(p)[0]))
        ax.imshow(img, cmap='gray'); ax.set_title(cls); ax.axis('off')
plt.show()
```
